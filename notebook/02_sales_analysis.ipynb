```python
import pandas as pd

# 1. 读取刚才清洗好的干净数据
# 注意：我们要读 csv 文件，不再是 excel 了
df = pd.read_csv("../data/cleaned_orders.csv", parse_dates=['下单时间'])

print(f"✅ 成功加载清洗后的数据: {df.shape[0]} 行")

# ==========================================
# 2. 计算核心指标 (KPI)
# ==========================================

# A. 总销售额 (GMV) - 这里用实收金额 (付款金额)
total_sales = df['付款金额'].sum()

# B. 总订单数
total_orders = df['订单号'].nunique() # 用 nunique 去重统计，更严谨

# C. 客单价 (Average Order Value) = 总销售额 / 总订单数
avg_order_value = total_sales / total_orders

print("\n📊 === 核心销售战报 ===")
print(f" 总销售额 (GMV): ¥{total_sales:,.2f}")
print(f"📦 总订单数: {total_orders:,} 单")
print(f"💵 客单价 (AOV): ¥{avg_order_value:.2f}")

# ==========================================
# 3. 平台业绩分析 (哪个渠道最赚钱？)
# ==========================================

# 按 '平台类型' 分组，计算销售额和订单量
platform_stats = df.groupby('平台类型').agg({
    '付款金额': 'sum',      # 销售额求和
    '订单号': 'nunique',    # 订单数去重计数
    '用户名': 'nunique'     # 购买用户数去重计数
}).rename(columns={
    '付款金额': '总销售额',
    '订单号': '订单量',
    '用户名': '购买用户数'
})

# 计算每个平台的客单价
platform_stats['客单价'] = platform_stats['总销售额'] / platform_stats['订单量']

# 按销售额从高到低排序
platform_stats = platform_stats.sort_values(by='总销售额', ascending=False)

print("\n🏆 === 各平台业绩排名 ===")
print(platform_stats)

# ==========================================
# 4. 优惠力度分析 (刚才造的新字段派上用场了)
# ==========================================

total_discount = df['优惠金额'].sum()
discount_rate = total_discount / (df['订单金额'].sum())

print(f"\n🎁 === 优惠分析 ===")
print(f"总优惠金额: ¥{total_discount:,.2f}")
print(f"整体折扣率: {discount_rate:.2%}") # 比如 0.05 就是 5%
```

```python
import pandas as pd
```

```python
# 读取数据
df = pd.read_excel("../data/datas.xlsx")
print(f"✅ 加载数据: {df.shape[0]} 行")
print(f"列名: {df.columns.tolist()}")
```

```python
# ==========================================
# 商品分析：三个维度看"最好"
# ==========================================

# 📦 维度1：按「订单数」排名（被下单的次数）
product_orders = df.groupby('商品编号')['订单号'].nunique().sort_values(ascending=False)

print("\n🏆 订单数 TOP10 商品：")
print(product_orders.head(10))
```

```python
# 💰 维度2：按「销售额」排名（带来的收入）
product_sales = df.groupby('商品编号')['付款金额'].sum().sort_values(ascending=False)

print("\n💎 销售额 TOP10 商品（按金额）：")
print(product_sales.head(10))
```

```python
# 📊 维度3：按「用户数」排名（购买人数）
product_users = df.groupby('商品编号')['用户名'].nunique().sort_values(ascending=False)

print("\n👥 购买人数 TOP10 商品：")
print(product_users.head(10))
```

```python
# ==========================================
# 商品贡献度分析（二八法则）
# ==========================================

total_sales = df['付款金额'].sum()

product_sales_df = df.groupby('商品编号')['付款金额'].sum().reset_index()
product_sales_df['占比'] = product_sales_df['付款金额'] / total_sales
product_sales_df = product_sales_df.sort_values('占比', ascending=False)

product_sales_df['累计占比'] = product_sales_df['占比'].cumsum()

print("\n📊 商品贡献度分析（前10名）：")
print(product_sales_df[['商品编号', '付款金额', '占比', '累计占比']].head(10))
```

```python
# 找出贡献 80% 销售额的商品数量
top_80_count = (product_sales_df['累计占比'] <= 0.8).sum()
print(f"\n🎯 二八法则验证：前 {top_80_count} 个商品贡献了 80% 的销售额")
```

```python
# ==========================================
# 用户分析
# ==========================================

# 用户消费金额排名
user_spending = df.groupby('用户名')['付款金额'].sum().sort_values(ascending=False)

print("\n💰 消费金额 TOP10 用户：")
print(user_spending.head(10))
```

```python
# 用户订单数排名
user_orders = df.groupby('用户名')['订单号'].nunique().sort_values(ascending=False)

print("\n📦 订单数 TOP10 用户：")
print(user_orders.head(10))
```

```python
# 保存分析结果
product_orders.to_csv("../output/04_product_orders.csv", encoding='utf-8-sig')
product_sales.to_csv("../output/04_product_sales.csv", encoding='utf-8-sig')
user_spending.to_csv("../output/04_user_spending.csv", encoding='utf-8-sig')
print("\n✅ 分析结果已保存到 output 文件夹")
```

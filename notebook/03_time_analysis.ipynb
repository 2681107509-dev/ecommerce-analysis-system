```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 读取清洗后的数据
df = pd.read_csv("../data/cleaned_orders.csv", parse_dates=['下单时间'])

# ==========================================
# 2. 按「天」分析销售趋势
# ==========================================

# 先把下单时间转成日期（去掉时分秒）
df['日期'] = df['下单时间'].dt.date

# 按日期分组，计算每天的总销售额
daily_sales = df.groupby('日期')['付款金额'].sum()

print("📅 每日销售额（前10天）：")
print(daily_sales.head(10))

# ==========================================
# 3. 按「小时」分析下单高峰
# ==========================================

# 提取小时（0-23）
df['小时'] = df['下单时间'].dt.hour

# 按小时分组，统计订单量
hourly_orders = df.groupby('小时')['订单号'].nunique()

print("\n⏰ 每小时订单量：")
print(hourly_orders)

# ==========================================
# 4. 按「星期几」分析
# ==========================================

# 提取星期几（Monday, Tuesday...）
df['星期'] = df['下单时间'].dt.day_name()

# 按星期分组，计算销售额
# 注意：day_name() 返回英文，可以手动映射成中文（可选）
weekday_sales = df.groupby('星期')['付款金额'].sum()

# 按星期顺序排序（不是字母顺序）
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_sales = weekday_sales.reindex(weekday_order)

print("\n🗓️ 每周销售额分布：")
print(weekday_sales)
# 保存分析结果到 output 文件夹
daily_sales.to_csv("../output/daily_sales.csv", encoding='utf-8-sig')
hourly_orders.to_csv("../output/hourly_orders.csv", encoding='utf-8-sig')
weekday_sales.to_csv("../output/weekday_sales.csv", encoding='utf-8-sig')
print("✅ 分析结果已保存到 output 文件夹")
```
